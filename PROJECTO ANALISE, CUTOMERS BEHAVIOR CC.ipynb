```python
#Creat a group column age_group

Labels = ["Young Adults", "Adult", "Middle-age", "Senior"]
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)

#df['age'] → pega a coluna de idades
#pd.qcut(..., q=4) → divide os dados em 4 partes com o mesmo número de pessoas
#labels=labels → dá nomes a cada grupo
#Resultado → cria a coluna age_group
```

```python
df[['age','age_group']].head(15)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

#Pega a coluna frequency_of_purchases do DataFrame df
#Usa .map() para substituir cada valor de texto pelo número correspondente do dicionário
#Cria (ou sobrescreve) uma nova coluna chamada purchase_frequency_days
```
